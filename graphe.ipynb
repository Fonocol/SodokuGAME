```python
import matplotlib.pyplot as plt
from math import exp

# Définition des fonctions de difficulté
def f(x):
    return 30 + 30 * (1 - exp(-x/100))

def g(x):
    return 20 + 30 * (1 - exp(-x/3))

def h(x):
    return 5 + 25 * (1 - exp(-x/3))

x0 = [i for i in range(100)]
y0 = [f(i) for i in x0]
y1 = [g(i) for i in x0]
y2 = [h(i) for i in x0]

# Trouver le point où chaque courbe atteint 95% de sa valeur maximale
x0_95_percent = next(x for x, y in zip(x0, y0) if y >= 0.95 * max(y0))
x1_95_percent = next(x for x, y in zip(x0, y1) if y >= 0.95 * max(y1))
x2_95_percent = next(x for x, y in zip(x0, y2) if y >= 0.95 * max(y2))

plt.plot(x0, y0, label="Difficult")
plt.plot(x0, y1, label="Medium")
plt.plot(x0, y2, label="Easy")

# Ajouter des marqueurs pour les points où chaque courbe atteint 95% de sa valeur maximale
plt.scatter(x0_95_percent, 0.95 * max(y0), color='blue', marker='o', label="95% Difficult")
plt.scatter(x1_95_percent, 0.95 * max(y1), color='red', marker='o', label="95% Medium")
plt.scatter(x2_95_percent, 0.95 * max(y2), color='green', marker='o', label="95% Easy")

# Ajouter des lignes verticales pour connecter les points 95%
plt.plot([x0_95_percent, x0_95_percent], [0, 0.95 * max(y0)], linestyle='--', color='blue')
plt.plot([x1_95_percent, x1_95_percent], [0, 0.95 * max(y1)], linestyle='--', color='red')
plt.plot([x2_95_percent, x2_95_percent], [0, 0.95 * max(y2)], linestyle='--', color='green')

plt.plot([0,x0_95_percent], [0.95 * max(y0),0.95 * max(y0)], linestyle='--', color='blue')
plt.plot([0,x1_95_percent], [0.95 * max(y1),0.95 * max(y1)], linestyle='--', color='red')
plt.plot([0,x2_95_percent], [0.95 * max(y2),0.95 * max(y2)], linestyle='--', color='green')

plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données depuis le fichier CSV
data = pd.read_csv("C:/Users/user/Desktop/lerning/SodokuGAME/sodoku_doc/data.csv", sep=";")


# Exemple de création d'un graphique à secteurs pour le niveau
plt.figure(figsize=(8, 6))
data['LEVEL'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Répartition des niveaux')
plt.ylabel('')
plt.show()

# Exemple de création d'un graphique à barres empilées pour le pays et le niveau
pivot_table = data.pivot_table(index='PAYS', columns='LEVEL', aggfunc='size', fill_value=0)
pivot_table.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('PAYS')
plt.ylabel('Nombre d\'individus')
plt.title('Répartition des niveaux par pays')
plt.xticks(rotation=45)
plt.show()
```
